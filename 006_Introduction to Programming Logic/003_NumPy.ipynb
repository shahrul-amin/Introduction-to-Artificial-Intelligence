## Broadcasting

```python
lst[0]=0
```

```python
lst
```

```python
lst[0:2]=0
```

```python
ar1=np.arange(1,10).reshape(3,3)
```

```python
ar1
```

```python
ar1[1:3,1:3]=0
```

```python
ar1
```

```python
#broadcasting
ar1[0:2,0]=200
```

```python
ar1
```

## Avoid Aliasing

```python
AAR1=np.arange(1,10).reshape(3,3)
AAR1
```

```python
AAR1_Slice=AAR1[1:3,1:3]
AAR1_Slice
```

```python
AAR1_Slice[:,:]=0
```

```python
AAR1_Slice
```

```python
AAR1
```

```python
AAR1=np.arange(1,10).reshape(3,3)
AAR1
```

```python
AAR1_Slice=AAR1.copy()[1:3,1:3]
```

```python
AAR1_Slice
```

```python
AAR1_Slice[:]=0
AAR1_Slice
```

```python
AAR1
```

```python
AAR1_Slice
```
